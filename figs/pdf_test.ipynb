```python
import matplotlib.pyplot as plt
import numpy as np

#plt.style.available       
#plt.style.use("ggplot") 
```

## Drift + Variation

```python
# Retention + variation
temperature = [300, 310, 320, 330, 340, 350, 360, 370, 380, 390, 400]
drift = [0.0, 0.013, 0.039, 0.065, 0.13, 0.2, 0.31, 0.35, 0.38, 0.4, 0.4]
acc_mnist = [98.49, 98.46, 98.30, 97.17, 86.44, 86.44, 19.54, 13.77, 11.27, 10.41, 10.41]
acc_cifar = [78.60, 78.04, 73.52, 65.68, 35.95, 17.02, 10.63, 10.20, 10.65, 10.57, 10.57]

#plt.figure(figsize = (6, 3))
plt.plot(drift, acc_mnist, '-s', label="MNIST")
plt.plot(drift, acc_cifar, '--^', label="CIFAR10")

plt.xlabel("Drift")
plt.ylabel("Accuracy")
plt.grid(linestyle='--')
plt.legend(frameon=False)
plt.tight_layout()
plt.rcParams['font.sans-serif'] = "Arial"
plt.rcParams['font.family'] = "sans-serif"
plt.rcParams['font.size'] = 10.
plt.savefig('drift.pdf')
plt.show()
```

## Vin Variation + G Variation

```python
# MNIST
# give vin a noise
vdd = [0.1, 0.3, 0.5, 0.7, 1, 2, 3]
acc_noise = [97.14, 98.39, 98.37, 98.37, 98.42, 98.40, 98.17]
acc_ideal = [97.37, 98.43, 98.55, 98.60, 98.49, 98.52, 98.42]
acc_irdrop = [97.39, 98.45, 98.55, 98.61, 98.50, 98.52, 98.42]

#plt.figure(figsize = (6, 3))
plt.plot(vdd, acc_ideal, '-s', label="without noise")
plt.plot(vdd, acc_noise, '--^', label="with noise")
plt.plot(vdd, acc_irdrop, ':x', label="with noise and ir_drop")

plt.xlabel("Vdd (V)")
plt.ylabel("Accuracy")
plt.ylim(95, 100)
plt.grid(linestyle='--')
plt.legend(frameon=False)
#plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', frameon=False)
plt.tight_layout()
plt.rcParams['font.sans-serif'] = "Arial"
plt.rcParams['font.family'] = "sans-serif"
plt.rcParams['font.size'] = 10.
plt.savefig('mnist_vdd.pdf')
plt.show()
```
